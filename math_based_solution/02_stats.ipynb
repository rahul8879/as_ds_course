```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets toss a coin 1 time and keep it in one variable
coin_toss = ['H', 'T']
# lets compute the probability of getting head or tail
total_events = len(coin_toss)
# lets say we want to compute the probability of getting head
favourable_events = coin_toss.count('H')
# lets compute the probability
probability_of_getting_head = favourable_events / total_events
probability_of_getting_tail= 1 - probability_of_getting_head
print(f"Probability of getting head is {probability_of_getting_head}")  


```

```python
# lest plot the probability distribution
outcomes = ['H', 'T']
probabilities = [0.5, 0.5]
plt.bar(outcomes, probabilities)
plt.xlabel('Outcomes')
plt.ylabel('Probabilities')
plt.title('Probability Distribution of a Fair Coin Toss')
plt.ylim(0, 1)
plt.show()
```
