```python
import pandas as pd

data = {
    'text': [
       'This is good',
       'This is bad',
       'This is okay',
       'This is average',
       'This is excellent',
       'This is worst'
    ],
    'label': [
        'positive',
        'negative',
        'neutral',
        'neutral',
        'positive',
        'negative'
    ]
}

df = pd.DataFrame(data)
df.head(2)

```

```python
p_y0 = df[df['label'] == 'positive'].shape[0] / df.shape[0]
p_y1 = df[df['label'] == 'negative'].shape[0] / df.shape[0]
p_y2 = df[df['label'] == 'neutral'].shape[0] / df.shape[0]

print(f"P(y=positive) = {p_y0}")
print(f"P(y=negative) = {p_y1}")
print(f"P(y=neutral) = {p_y2}")
```

```python
# create table for prior probabilities
prior_probs = pd.DataFrame({
    'label': ['positive', 'negative', 'neutral'],
    'probability': [p_y0, p_y1, p_y2]
})
prior_probs 
```

```python
# lets create vocabulary
vocab = set()
df['text'].apply(lambda x: vocab.update(x.split()))
vocab = list(vocab) 


def word_probabilities(vocab, df):
    word_probs = {}
    for word in vocab:
        p_word_given_positive = df[df['label'] == 'positive']['text'].str.contains(word).sum() / df[df['label'] == 'positive'].shape[0]
        p_word_given_negative = df[df['label'] == 'negative']['text'].str.contains(word).sum() / df[df['label'] == 'negative'].shape[0]
        p_word_given_neutral = df[df['label'] == 'neutral']['text'].str.contains(word).sum() / df[df['label'] == 'neutral'].shape[0]
        word_probs[word] = {'positive': p_word_given_positive, 'negative': p_word_given_negative, 'neutral': p_word_given_neutral}
    return word_probs

word_probs = word_probabilities(vocab, df)
word_probs
```

```python
new_text = "Rahul worst"
import numpy as np
def predict(text, prior_probs, word_probs):
    # Preprocess the text
    words = text.split()
    # Calculate the log posterior for each class
    log_posteriors = {}
    for label in prior_probs['label']:
        log_posteriors[label] = np.log(prior_probs[prior_probs['label'] == label]['probability'].values[0])
        for word in words:
            if word in word_probs:
                log_posteriors[label] += np.log(word_probs[word][label])
    # Return the class with the highest posterior
    return max(log_posteriors, key=log_posteriors.get)

prediction = predict(new_text, prior_probs, word_probs)
prediction
```
